# Importing and Cleaning final_consumption.csv file

```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# Import Data file
consumption_exp = pd.read_csv("../Resources/final_consumption_expenditure.csv")
consumption_df = pd.DataFrame(consumption_exp)
consumption_df
```

```python
# Droping the top 3 rows
consumption_df = consumption_df.drop(range(0,3))
consumption_df
```

```python
# Setting first column as Index
new_header = consumption_df.iloc[0]
consumption_df = consumption_df[1:]
consumption_df.columns = new_header
consumption_df
```

```python
# Name of the columns
consumption_df.columns
```

```python
# Type of columns
consumption_df.dtypes
```

```python
# Dropping the 2021 column
consumption_df = consumption_df.drop(columns = consumption_df.columns[2:45], inplace = False, axis = 1)
consumption_df
```

```python
consumption_df.count()
```

```python
# Dropping the 2021 column
consumption_df = consumption_df.drop(columns = consumption_df.columns[21:], inplace = False, axis = 1)
consumption_df
```

```python
# Setting the Index
consumption_df = consumption_df.set_axis(['Country_Name', 'Country_Code', 'con_2001', 'con_2002', 'con_2003', 
                                          'con_2004', 'con_2005', 'con_2006', 'con_2007', 'con_2008', 'con_2009',
                                          'con_2010', 'con_2011', 'con_2012', 'con_2013', 'con_2014', 'con_2015',
                                          'con_2016', 'con_2017', 'con_2018', 'con_2019'], axis = 1)
consumption_df
```

```python
consumption_df = consumption_df.dropna()
consumption_df
```

```python
# Resetting the index to Zero
consumption_df = consumption_df.reset_index(drop=True)
consumption_df
```

# Importing and merging Country-codes.csv

```python
# Import Data file
country = pd.read_csv("../Resources/country-codes.csv")
country_df = pd.DataFrame(country)
country_df
```

```python
country_df = country_df.set_axis(['Country Name', 'Country Alpha-2 Code', 'Country_Code', 'Numeric code',
                                 'ISO 3166-2'], axis = 1)
country_df
```

```python
new_df = pd.merge(consumption_df, country_df, on= 'Country_Code', suffixes = ['', '_country' ])
new_df 
```

# Importing and Merging region_income.csv

```python
# Import Data file
income_group = pd.read_csv("../Resources/countries_region_income.csv")
income_group_df = pd.DataFrame(income_group)
income_group_df
```

```python
income_group_df = income_group_df.rename(columns = {'Country Code': 'Country_Code'})
income_group_df
```

```python
new_consumption_df = pd.merge(new_df, income_group_df, on= 'Country_Code', suffixes = ['_new', '_income'])
new_consumption_df 
```

```python

new_consumption_df = new_consumption_df.drop(['Country Name', 'Country Alpha-2 Code', 'Numeric code', 'ISO 3166-2', 'Region',
                                    'SpecialNotes', 'TableName'], axis = 1)
new_consumption_df
```

```python
new_consumption_df = new_consumption_df.rename(columns = {'IncomeGroup': 'Income_Group'})
new_consumption_df
```

# Cleaned Consumption DataFrame

```python
new_consumption_df = new_consumption_df.round(2)
new_consumption_df
```

```python
yearwise_consumption = new_consumption_df.mean(axis =0, numeric_only=None)
yearwise_consumption
```

# Yearwise Consumption DataFrame

```python
annual_consumption_df = pd.DataFrame({"consumption": yearwise_consumption})
annual_consumption_df.round(2)
```

```python
annual_consumption_df = annual_consumption_df.round(2)
annual_consumption_df
```

```python
# Create the output file (CSV).
output_data_file = "../cleaned_data/annual_consumption.csv"
# Export the City_Data into a CSV.
annual_consumption_df.to_csv(output_data_file, index = True)
```
