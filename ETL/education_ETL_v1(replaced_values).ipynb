## Number of Universities

```python
#initial imports
university_data = pd.read_csv(webscraped_universities)
university_df = pd.DataFrame(university_data)
university_df
```

```python
#handle duplicated index
university_df = university_df.set_index('Unnamed: 0', drop =True)
university_df = university_df.rename_axis(None)
university_df
```

```python
#drop null rows
university_df = university_df.dropna()
university_df
```

```python
#drop duplicate rows
university_df = university_df.drop_duplicates()
university_df
```

```python
#convert 2 letter country code to caps to match education_df
university_df["Country"] = university_df["Country"].str.upper()
university_df
```

```python
#count universities and crete new df
university_count = university_df.groupby(["Country"]).count()
university_count
```

```python
#create df 
university_count_df = pd.DataFrame(university_count.reset_index().rename(columns={"Country": "Alpha-2 code", "University Name" : "university_count"}))
university_count_df
```

```python
#merge df with university_df
education_df = education_df.merge(university_count_df, how="left",on=["Alpha-2 code"])
education_df
```

```python
#drop Alpha-2 code column
education_df = education_df.drop(education_df.columns[21], axis=1)
education_df
```

```python
education_df.isna().count()
```

```python
#fill null values with 0
education_df = education_df.fillna(0)
education_df
```

```python
# education_df.isna().count()
education_df = education_df.dropna()
education_df
```

```python
#convert university count to numeric
#solution found at https://stackoverflow.com/questions/36825972/utf-8-codec-cant-decode-byte-0x80
education_df = education_df[~education_df['university_count'].isnull()]
education_df[['university_count']] = education_df[['university_count']].astype(int)
education_df
```

```python
education_df.isna().sum()
```

## Tuition

```python
tuition_data = pd.read_csv(tuition_filepath)
tuition_df = pd.DataFrame(tuition_data)
tuition_df
```

```python
free_with_restrictions_df = tuition_df[tuition_df['details'].str.contains("Free")]
free_with_restrictions_df
```

```python
#simplify tuition to a boolean
#create a new "free" column
free_with_restrictions_df["Simplified_Tuition"] = "Free"

#drop unnecessary columns
free_with_restrictions_df = free_with_restrictions_df.drop(free_with_restrictions_df.columns[1:3], axis=1)
free_with_restrictions_df
```

```python
#merge df with region_income_df
education_df = education_df.merge(free_with_restrictions_df, how="left", left_on=["Country Name"], right_on=["country"])
education_df.head(20)
```

```python
#drop country column
education_df = education_df.drop(education_df.columns[23], axis=1)
education_df
```

```python
# education_df.loc[education_df['university_count'] == '0', 'Simplified_Tuition'] = 'None'
education_df["Simplified_Tuition"] = np.where(education_df["university_count"]> 0, "Standard", "NA")
education_df
```

```python
#clean Up DataFrame
#replace column names
education_df = education_df.set_axis(['Country_Name', 'Country_Code', '2001', '2002','2003','2004','2005','2006',
                                            '2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017',
                                            '2018','2019', 'Income_Group','University_Count', 'Tuition'], axis = 1)
education_df.head()
```
