## Region Income


```python
#merge in dataframe for consistency in country names across various tables
#read in csv
region_income = pd.read_csv(countries_region_income)

#create DataFrame
region_income_df = pd.DataFrame(region_income)
region_income_df
```

```python
#merge df with region_income_df
df = df.merge(region_income_df, how="left", left_on=["country_code"], right_on=["Country Code"])
df.head(20)
```

```python
#drop country column
df = df.drop(["Country Code", "SpecialNotes", "TableName"], axis=1)
df
```

```python
#b. rename new Country name Column
df = df.rename({"Region": "region", "IncomeGroup": "income_group"}, axis=1)
df
```

```python
#simplify income category for machine learning
df['income_group'] = df["income_group"].str.replace("Upper middle income", "High income", case=False)
df['income_group'] = df["income_group"].str.replace("Lower middle income", "Low income", case=False)
df
```

```python
#export to csv
df.to_csv("Resources/cleaned_datasets/clean_GDP_per_capita.csv", index=False)
```
