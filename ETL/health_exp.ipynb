# Importing and Cleaning Health_expenditure.csv file 

```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# Import Data file
health_exp = pd.read_csv("../Resources/health_expenditure.csv")
health_exp_df = pd.DataFrame(health_exp)
health_exp_df
```

```python
# Droping the top 3 rows
health_exp_df = health_exp_df.drop(range(0,3))
health_exp_df
```

```python
# Setting first column as Index
new_header = health_exp_df.iloc[0]
health_exp_df = health_exp_df[1:]
health_exp_df.columns = new_header
health_exp_df
```

```python
# Name of the columns
health_exp_df.columns
```

```python
# Type of columns
health_exp_df.dtypes
```

```python
# Dropping the columns from year 1960 to 2000
health_exp_df = health_exp_df.drop(health_exp_df.columns[2:45], inplace = False, axis = 1)
health_exp_df
```

```python
# Dropping the 2021 column
health_exp_df = health_exp_df.drop(columns = health_exp_df.columns[21:23], inplace = False, axis = 1)
health_exp_df
```

```python
health_exp_df.columns
```

```python
# Setting the Index
health_exp_df = health_exp_df.set_axis(['Country_Name', 'Country_Code', '2001', '2002', '2003', '2004', '2005',         
                                        '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', 
                                        '2015', '2016', '2017', '2018', '2019'], axis = 1)
health_exp_df
```

```python
health_exp_df.count()
```

```python
health_exp_df = health_exp_df.dropna()
health_exp_df
```

```python
# Resetting the index to Zero
health_exp_df = health_exp_df.reset_index(drop=True)
health_exp_df
```

```python
for country in health_exp_df['Country_Name'].unique():
    print(country)
```

# Importing and merging Country-codes.csv

```python
# Import Data file
country = pd.read_csv("../Resources/country-codes.csv")
country_df = pd.DataFrame(country)
country_df
```

```python
country_df = country_df.set_axis(['Country Name', 'Country Alpha-2 Code', 'Country_Code', 'Numeric code',
                                 'ISO 3166-2'], axis = 1)
country_df
```

```python
for country in country_df['Country Name'].unique():
    print(country)
```

```python
new_df = pd.merge(health_exp_df, country_df, on= 'Country_Code', suffixes = ['', '_country' ])
new_df 
```

```python
new_df.isnull().sum()
```

```python
for country in new_df['Country Name'].unique():
    print(country)
```

# Importing and Merging region_income.csv

```python
# Import Data file
income_group = pd.read_csv("../Resources/countries_region_income.csv")
income_group_df = pd.DataFrame(income_group)
income_group_df
```

```python
income_group_df = income_group_df.set_axis(['Country_Code', 'Region', 'IncomeGroup', 'SpecialNotes', 'TableName'], axis = 1)
country_df
```

```python
new_health_df = pd.merge(new_df, income_group_df, on= 'Country_Code', suffixes = ['_new', '_income'])
new_health_df 
```

```python

new_health_df = new_health_df.drop(['Country Name', 'Country Alpha-2 Code', 'Numeric code', 'ISO 3166-2', 'Region',
                                    'SpecialNotes', 'TableName'], axis = 1)
new_health_df
```

```python
new_health_df = new_health_df.rename(columns = {'IncomeGroup': 'Income_Group'})
new_health_df
```

# Cleaned DataFrame

```python
new_health_df = new_health_df.round(2)
new_health_df
```

```python
yearwise_health_df = new_health_df.mean(axis =0, numeric_only=None)
yearwise_health_df
```

# Yearwise Health DataFrame

```python
yearwise_health_df = yearwise_health_df.round(2)
yearwise_health_df
```
