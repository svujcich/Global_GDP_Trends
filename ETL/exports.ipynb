# Importing and Cleaning exports_of_goods&services.csv file

```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# Import Data file
exports = pd.read_csv("../Resources/exports_of_goods&services.csv")
exports_df = pd.DataFrame(exports)
exports_df
```

```python
# Droping the top 3 rows
exports_df = exports_df.drop(range(0,3))
exports_df
```

```python
# Setting first column as Index
new_header = exports_df.iloc[0]
exports_df = exports_df[1:]
exports_df.columns = new_header
exports_df
```

```python
# Dropping the columns from year 1960 to 2000
exports_df = exports_df.drop(exports_df.columns[2:45], inplace = False, axis = 1)
exports_df
```

```python
# Dropping the 2021 column
exports_df = exports_df.drop(columns = exports_df.columns[21:], inplace = False, axis = 1)
exports_df
```

```python
# Setting the Index
exports_df = exports_df.set_axis(['Country_Name', 'Country_Code', 'ex_2001', 'ex_2002', 'ex_2003', 'ex_2004', 
                                  'ex_2005', 'ex_2006', 'ex_2007', 'ex_2008', 'ex_2009', 'ex_2010', 'ex_2011', 
                                  'ex_2012', 'ex_2013', 'ex_2014', 'ex_2015', 'ex_2016', 'ex_2017', 'ex_2018',
                                  'ex_2019'], axis = 1)
exports_df
```

```python
exports_df.count()
```

```python
exports_df = exports_df.dropna()
exports_df
```

```python
# Resetting the index to Zero
exports_df = exports_df.reset_index(drop=True)
exports_df
```

# Importing and merging Country-codes.csv

```python
# Import Data file
country = pd.read_csv("../Resources/country-codes.csv")
country_df = pd.DataFrame(country)
country_df
```

```python
country_df = country_df.set_axis(['Country_Name', 'Country Alpha-2 Code', 'Country_Code', 'Numeric code',
                                 'ISO 3166-2'], axis = 1)
country_df
```

```python
new_exports_df = pd.merge(exports_df, country_df, on= 'Country_Code', suffixes = ['', '_country' ])
new_exports_df 
```

# Importing and Merging region_income.csv

```python
# Import Data file
income_group = pd.read_csv("../Resources/countries_region_income.csv")
income_group_df = pd.DataFrame(income_group)
income_group_df
```

```python
income_group_df = income_group_df.rename(columns = {'Country Code': 'Country_Code'})
```

```python
new_exports_df = pd.merge(new_exports_df, income_group_df, on= 'Country_Code', suffixes = ['_new', '_income'])
new_exports_df 
```

```python

new_exports_df = new_exports_df.drop(['Country_Name_country', 'Country Alpha-2 Code', 'Numeric code', 'ISO 3166-2', 'Region',
                                    'SpecialNotes', 'TableName'], axis = 1)
new_exports_df
```

```python
new_exports_df = new_exports_df.rename(columns = {'IncomeGroup': 'Income_Group'})
new_exports_df
```

```python
new_exports_df = new_exports_df.round(2)
new_exports_df
```

```python
new_exports_df.dtypes
```

```python
yearwise_exports = new_exports_df.mean(axis = 0, numeric_only = None)
yearwise_exports 
```

# Yearwise Consumption DataFrame

```python
annual_exports_df = pd.DataFrame({"exports": yearwise_exports})
annual_exports_df.round(2)
```

```python
annual_exports_df = annual_exports_df.round(2)
annual_exports_df
```

```python
# Create the output file (CSV).
output_data_file = "../cleaned_datasets/mean_data/annual_exports.csv"
# Export the City_Data into a CSV.
annual_exports_df.to_csv(output_data_file, index = True)
```
