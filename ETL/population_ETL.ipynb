```python
#initial imports
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
#files to load
population = ("Resources/population.csv")
country_code_file_path =("Resources/country-codes.csv") 
countries_region_income = ("Resources/countries_region_income.csv")
```

```python
#read into dataframe
population_data = pd.read_csv(population)
df = pd.DataFrame(population_data)
df
```

```python
#get header rows
#a. create empty list for headers
headers = []
```

```python
#b. drop first 3 rows with no data
population_df = df.iloc[3:, :]

#c. use new first row as headers
headers = population_df.iloc[0]
population_df = population_df[1:]
population_df.columns = headers

population_df.head(10)
```

```python
#drop columns that will not be of use
population_df = population_df.drop(population_df.columns[2:45], axis=1)
population_df = population_df.drop(population_df.columns[21:23], axis=1)
population_df
```

```python
#clean Up DataFrame
#replace column names
population_df = population_df.set_axis(['Country_Name', 'Country_Code', '2001', '2002','2003','2004','2005','2006',
                                            '2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017',
                                            '2018','2019'], axis = 1)
population_df.head()
```

```python
print(population_df.dtypes)
print(population_df.shape)
```

## Merge with country codes list

```python
#create df
#read in df for with country codes
country_codes = pd.read_csv(country_code_file_path)
country_codes_df = pd.DataFrame(country_codes)
country_codes_df
```

```python
#rename country column in country_codes_df
country_codes_df = country_codes_df.rename({"English short name lower case": "Country_Name", "Alpha-3 code":"Country_Code"}, axis=1)
country_codes_df
```

```python
#merge df with region_income_df
population_df = population_df.merge(country_codes_df, how="left", left_on=["Country_Code"], right_on=["Country_Code"])
population_df 
```

```python
print(population_df.columns)
```

```python
#check null values
population_df.isnull().sum()
```

```python
#Check that null values are not actually countries
new_null_rows = population_df[population_df.isna().any(axis=1)]
new_null_rows_df = pd.DataFrame(new_null_rows)
new_null_rows_df["Country_Name_x"]
```

```python
#drop rows that are not countries
population_df = population_df.dropna()
population_df
```

```python
#clean up data frame

#a. drop country name row from GDP_growth_df
population_df = population_df.drop(['Country_Name_x', 'Alpha-2 code', 'Numeric code', 'ISO 3166-2'], axis=1)

#b. rename new Country name Column
population_df = population_df.rename({"Country_Name_y": "Country_Name"}, axis=1)
```

```python
#c. list column names 
columns = list(population_df.columns.values)
print(columns)
```

```python
#d.reorganize columns
population_df = population_df.loc[:,["Country_Name",'Country_Code', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']]
population_df
```

## Region Income

```python
#read in csv
region_income = pd.read_csv(countries_region_income)

#create DataFrame
region_income_df = pd.DataFrame(region_income)
region_income_df
```

```python
#merge df with region_income_df
population_df = population_df.merge(region_income_df, how="left", left_on=["Country_Code"], right_on=["Country Code"])
population_df 
```

```python
#check null values
population_df.isnull().sum()
```

```python
#drop unnecessary columns
#a. list columns
columns = list(population_df.columns.values)
print(columns)
```

```python
#b. drop columns
population_df = population_df.drop(['Region', 'SpecialNotes', 'TableName', 'Country Code'], axis=1)
population_df
```

```python
#b. rename  income group column
population_df = population_df.rename({"IncomeGroup": "Income_Group"}, axis=1)
population_df
```

```python
#replace column names
population_df = population_df.set_axis(['country_name', 'country_code', 'p_2001', 'p_2002','p_2003','p_2004','p_2005','p_2006',
                                            'p_2007','p_2008','p_2009','p_2010','p_2011','p_2012','p_2013','p_2014','p_2015','p_2016','p_2017',
                                            'p_2018','p_2019', 'income_group'], axis = 1)
population_df
```

```python
#check null values
population_df.isnull().sum()
```

```python
#drop null row
population_df = population_df.dropna()
population_df.isnull().sum()
```

```python
#Simplify income groups for machine learning
population_df['income_group'] = population_df["income_group"].str.replace("Upper middle income", "High income", case=False)
population_df['income_group'] = population_df["income_group"].str.replace("Lower middle income", "Low income", case=False)
population_df
```

```python
#export to csv
population_df.to_csv("Resources/cleaned_datasets/clean_population.csv",index=False)
```

## Top 10, bottom 10, mean

```python
#check data types
population_df.dtypes
```

```python
#find top 10 countries
top_10_countries = population_df.sort_values(by="p_2019",ascending= False).head(10)
top_10_countries = top_10_countries.reset_index(drop=True)


print("Countries With the Largest Population in 2019")
top_10_countries
```

```python
#Find the bottom ten countries in 2019.
#find 10 countries with cmallest gdp growth
bottom_10_countries = population_df.sort_values(by="p_2019",ascending= True).head(10)
#reset index
bottom_10_countries = bottom_10_countries.reset_index(drop=True)

print('Countries With the Smallest Population in 2019')
bottom_10_countries
```

```python
#create df of mean global population
#create varible
mean_population_by_year = population_df.mean(axis = 0, numeric_only=None)
#create df with varible
mean_population_df = pd.DataFrame(mean_population_by_year)
#turn axis into column
mean_population_df.reset_index(inplace=True)
#set axis name
mean_population_df = mean_population_df.set_axis(['Year','Mean'], axis=1)

mean_population_df
```

```python
#export to csv
mean_population_df.to_csv("Resources/cleaned_datasets/mean_population.csv", index=False)
```
